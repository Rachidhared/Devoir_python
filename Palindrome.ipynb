```python

def is_palindrome_while(input_string: str) -> bool:

    nouveau = ''.join(e for e in input_string if e.isalnum()).lower()

    G = 0
    D = len(nouveau) - 1

    while G < D:
        if nouveau[G] != nouveau[D]:
            return False
        G += 1
        D -= 1

    return True

input_string = input("Entrez un mot ou une phrase: ")

if is_palindrome_while(input_string):
    print(f"'{input_string}' est un palindrome.")
else:
    print(f"'{input_string}' n'est pas un palindrome.")






```

```python
def is_palindrome_for(input_string: str) -> bool:

    nouveau = ''.join(c for c in input_string if c.isalnum()).lower()

    G = 0
    D = len(nouveau) - 1

    for _ in range(len(nouveau) // 2):
        if nouveau[G] != nouveau[D]:

            return False
        G += 1
        D -= 1

    return True

input_string = input("Entrez un mot ou une phrase: ")

if is_palindrome_for(input_string):
    print(f"'{input_string}' est un palindrome.")
else:
    print(f"'{input_string}' n'est pas un palindrome.")
```
