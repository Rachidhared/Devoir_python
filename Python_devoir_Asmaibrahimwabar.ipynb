```python
Exercice 1 : Concevez Votre Propre Jeu de Chasse au Trésor

import random

def afficher_plateau(plateau):
    for ligne in plateau:
        print(" ".join(ligne))

def creer_plateau(n, m):
    return [["*" for _ in range(m)] for _ in range(n)]

def demander_emplacement(message, n, m):
    while True:
        try:
            ligne = int(input(f"{message} (entre 1 et {n}): ")) - 1
            colonne = int(input(f"Colonne (entre 1 et {m}): ")) - 1
            if 0 <= ligne < n and 0 <= colonne < m:
                return ligne, colonne
            else:
                print("Erreur : emplacement en dehors des limites du plateau.")
        except ValueError:
            print("Erreur : veuillez entrer un numéro valide.")

def ajouter_piege(plateau, ligne_tresor, colonne_tresor):
    n = len(plateau)
    m = len(plateau[0])
    while True:
        ligne_piege = random.randint(0, n - 1)
        colonne_piege = random.randint(0, m - 1)
        if (ligne_piege != ligne_tresor or colonne_piege != colonne_tresor):
            plateau[ligne_piege][colonne_piege] = "X"
            break

def jeu():
    n = int(input("Nombre de lignes du plateau: "))
    m = int(input("Nombre de colonnes du plateau: "))

    plateau = creer_plateau(n, m)
    afficher_plateau(plateau)

    ligne_tresor, colonne_tresor = demander_emplacement("Entrez l'emplacement du trésor", n, m)
    plateau[ligne_tresor][colonne_tresor] = "T"

    ajouter_piege(plateau, ligne_tresor, colonne_tresor)

    afficher_plateau(plateau)

    while True:
        print("\nEssayez de trouver le trésor!")
        ligne_devine, colonne_devine = demander_emplacement("Entrez l'emplacement que vous pensez", n, m)
        
        if ligne_devine == ligne_tresor and colonne_devine == colonne_tresor:
            print("Félicitations! Vous avez trouvé le trésor!")
            break
        elif plateau[ligne_devine][colonne_devine] == "X":
            print("Vous êtes tombé dans le piège! Game Over.")
            break
        else:
            print("Ce n'est pas le bon endroit. Essayez encore.")

if __name__ == "__main__":
    jeu()
```

```python
Exercice 2 : Création d'un Calculateur d'IMC en Utilisant des Fonctions
def calculate_bmi(weight, height):
    if height <= 0:
        raise ValueError("La taille doit être un nombre positif.")
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def afficher_categorie_bmi(bmi):
    if bmi < 18.5:
        return "Insuffisance pondérale"
    elif 18.5 <= bmi < 25:
        return "Poids normal"
    elif 25 <= bmi < 30:
        return "Surpoids"
    else:
        return "Obésité"

def main():
    print("Choisissez les unités pour le poids et la taille :")
    print("1. Kilogrammes et mètres")
    print("2. Livres et pouces")
    choix = input("Entrez 1 ou 2 : ")

    if choix == '1':
        poids = float(input("Entrez votre poids en kilogrammes : "))
        taille = float(input("Entrez votre taille en mètres : "))
    elif choix == '2':
        poids_livres = float(input("Entrez votre poids en livres : "))
        taille_pouces = float(input("Entrez votre taille en pouces : "))
        
        poids = poids_livres / 2.20462
        taille = taille_pouces / 39.3701
    else:
        print("Choix non valide. Le programme va se terminer.")
        return

    try:
       
        bmi = calculate_bmi(poids, taille)
        print(f"Votre IMC est : {bmi}")
        categorie = afficher_categorie_bmi(bmi)
        print(f"Catégorie de santé : {categorie}")
    except ValueError as e:
        print(e)

if __name__ == "__main__":
    main()
```

```python
import re

def nettoyer_chaine(input_string: str) -> str:
    return re.sub(r'[^a-zA-Z0-9]', '', input_string).lower()

def is_palindrome_while(input_string: str) -> bool:
    nettoye = nettoyer_chaine(input_string)
    gauche, droite = 0, len(nettoye) - 1
    
    while gauche < droite:
        if nettoye[gauche] != nettoye[droite]:
            return False
        gauche += 1
        droite -= 1
    
    return True

def is_palindrome_for(input_string: str) -> bool:
    nettoye = nettoyer_chaine(input_string)
    longueur = len(nettoye)
    
    for i in range(longueur // 2):
        if nettoye[i] != nettoye[longueur - i - 1]:
            return False
    
    return True

def tests():
    exemples = [
        "A man, a plan, a canal, Panama",
        "No 'x' in Nixon",
        "Hello, World!",
        "Madam, in Eden, I'm Adam",
        "Python"
    ]

    for phrase in exemples:
        print(f"'{phrase}' est un palindrome (while) ? {is_palindrome_while(phrase)}")
        print(f"'{phrase}' est un palindrome (for) ? {is_palindrome_for(phrase)}")

if __name__ == "__main__":
    tests()
```
