**Création D'un Calculateur d'IMC**

```python
def calculate_bmi(poids,taille):
  IMC=poids/taille
  return IMC

print("*** Menu \n 1 Poids en kg et Taille en m²\n 2 Poids en livres et Taille en pouces")
choix=int(input("Veuillez choisir 1 ou 2 "))
if choix==1:
  poids=float(input("Veuillez saisir votre Poids en kg :"))
  taille=float(input("Veuillez saisir votre taille m² :"))
  if calculate_bmi(poids,taille)<18.5:
    print("Insuffissance ponderale")
  elif 18.5<calculate_bmi(poids,taille)<24.9:
    print("Poids normal")
  elif 25<calculate_bmi(poids,taille)<29.9:
    print("Surpoids")
  else :
    print("Obésité")
elif choix==2:
  poids=float(input("Veuillez saisir votre Poids en lb :"))
  taille=float(input("Veuillez saisir votre taille inch :"))
  if (calculate_bmi(poids,taille)*(2.20462/39.3701))<18.5:
    print("Insuffissance ponderale")
  elif 18.5<(calculate_bmi(poids,taille)*(2.20462/39.3701))<24.9:
    print("Poids normal")
  elif 25<(calculate_bmi(poids,taille)*(2.20462/39.3701))<29.9:
    print("Surpoids")
  else :
    print("Obésité")
else :
  print("Veuillez saisir 1 ou 2 ")
```
