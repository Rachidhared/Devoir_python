```python
import random

def creer_grille(rows, cols):
    return [['-' for _ in range(cols)] for _ in range(rows)]

def afficher_grille(grille):  
    for ligne in grille:
        print(' '.join(ligne))
    print()

def placer_tresor(grille, row, col):
    grille[row][col] = 'T'

def placer_piege(grille, row_tresor, col_tresor):
    rows = len(grille)
    cols = len(grille[0])
    while True:
        row_piege = random.randint(0, rows - 1)
        col_piege = random.randint(0, cols - 1)
        if (row_piege, col_piege) != (row_tresor, col_tresor):
            grille[row_piege][col_piege] = 'X'
            break

def verifier_devine(grille, row, col):
    if grille[row][col] == 'T':
        return "Bravo ! Vous avez trouvé le trésor !"
    elif grille[row][col] == 'X':
        return "Oups ! Vous êtes tombé dans un piège !"
    else:
        return "Pas encore. Essayez ailleurs !"

def jeu():
    rows = int(input("Entrez le nombre de lignes de la grille : "))
    cols = int(input("Entrez le nombre de colonnes de la grille : "))

    grille = creer_grille(rows, cols)

    while True:
        try:
            row_tresor = int(input(f"Entrez la ligne pour cacher le trésor (0 à {rows - 1}): "))
            col_tresor = int(input(f"Entrez la colonne pour cacher le trésor (0 à {cols - 1}): "))
            if 0 <= row_tresor < rows and 0 <= col_tresor < cols:
                break
            else:
                print("Emplacement invalide. Veuillez réessayer.")
        except ValueError:
            print("Veuillez entrer des nombres valides.")

    placer_tresor(grille, row_tresor, col_tresor)

    placer_piege(grille, row_tresor, col_tresor)

    afficher_grille(grille)

    while True:
        try:
            row_devine = int(input(f"Devinez la ligne du trésor (0 à {rows - 1}): "))
            col_devine = int(input(f"Devinez la colonne du trésor (0 à {cols - 1}): "))
            if 0 <= row_devine < rows and 0 <= col_devine < cols:
                break
            else:
                print("Emplacement invalide. Veuillez réessayer.")
        except ValueError:
            print("Veuillez entrer des nombres valides.")

    resultat = verifier_devine(grille, row_devine, col_devine)
    print(resultat)

if __name__ == "__main__":
    jeu()
```
