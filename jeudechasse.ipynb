```python
def creer_plateau(n: int, m: int) -> list:
    """
    Crée un plateau de taille n x m rempli d'astérisques.
    """
    return [['*' for _ in range(m)] for _ in range(n)]

def afficher_plateau(plateau: list) -> None:
    """
    Affiche le plateau de jeu.
    """
    for ligne in plateau:
        print(' '.join(ligne))

def demander_position(plateau: list) -> None:
    """
    Demande à l'utilisateur de spécifier l'emplacement du trésor et place le trésor sur le plateau.
    """
    n = len(plateau)
    m = len(plateau[0])

    while True:
        try:
            ligne = int(input(f"Entrez le numéro de la ligne (0 à {n-1}): "))
            colonne = int(input(f"Entrez le numéro de la colonne (0 à {m-1}): "))

            if 0 <= ligne < n and 0 <= colonne < m:
                plateau[ligne][colonne] = 'T'
                break
            else:
                print("Erreur : Les coordonnées sont en dehors des dimensions du plateau. Essayez à nouveau.")
        except ValueError:
            print("Erreur : Veuillez entrer des nombres entiers valides.")

def jeu_chasse_au_tresor() -> None:
    """
    Fonction principale pour exécuter le jeu de chasse au trésor.
    """
    print("Bienvenue dans le jeu de chasse au trésor !")

    # Demander les dimensions du plateau
    while True:
        try:
            n = int(input("Entrez le nombre de lignes du plateau: "))
            m = int(input("Entrez le nombre de colonnes du plateau: "))

            if n > 0 and m > 0:
                break
            else:
                print("Erreur : Les dimensions doivent être des entiers positifs. Essayez à nouveau.")
        except ValueError:
            print("Erreur : Veuillez entrer des nombres entiers valides.")

    # Créer le plateau
    plateau = creer_plateau(n, m)

    # Demander la position du trésor
    demander_position(plateau)

    # Afficher le plateau
    print("\nPlateau de jeu :")
    afficher_plateau(plateau)

# Exécuter le jeu
jeu_chasse_au_tresor()
```
