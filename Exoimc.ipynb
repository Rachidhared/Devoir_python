```python
def calculate_bmi () :
  poids= float ( input ( "veuillez saisir votre poids : " ))
  taille= float ( input ( "veuillez saisir votre taille :" ))
  IMC=poids/ ( taille** 2 )
  return IMC
IMC = calculate_bmi ()
print ( f" Votre IMC est:  {IMC:.2f} ")

if  IMC <  18.5 :
  print ( "Insuffisance pondérale" )
elif 18.5  <= IMC <=  24.9 :
  print ( "Poids normaux" )
elif 25  <= IMC <=  29.9 :
  print ( "Surpoids" )
elif  IMC >=  30 : print ( "Obésité" )
```
