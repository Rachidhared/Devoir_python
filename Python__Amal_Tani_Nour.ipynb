```python
Exercice 1

import random

def afficher_plateau(plateau):
    for ligne in plateau:
        print(" ".join(ligne))

def creer_plateau(n, m):
    return [["*" for _ in range(m)] for _ in range(n)]

def demander_position(n, m, message=""):
    while True:
        try:
            x, y = map(int, input(message).split())
            if 0 <= x < n and 0 <= y < m:
                return x, y
            else:
                print(f"Position invalide. Veuillez entrer des nombres entre 0 et {n-1} pour les lignes et 0 et {m-1} pour les colonnes.")
        except ValueError:
            print("Veuillez entrer des nombres valides séparés par un espace.")

def placer_piege(plateau, tresor_pos):
    n, m = len(plateau), len(plateau[0])
    while True:
        piege_pos = (random.randint(0, n-1), random.randint(0, m-1))
        if piege_pos != tresor_pos:
            return piege_pos

def main():
    # Étape 1 : Demander la taille du plateau
    n = int(input("Entrez le nombre de lignes du plateau: "))
    m = int(input("Entrez le nombre de colonnes du plateau: "))

    # Étape 2 : Créer le plateau
    plateau = creer_plateau(n, m)

    # Étape 3 : Demander la position du trésor
    tresor_pos = demander_position(n, m, "Entrez la position du trésor (ligne colonne): ")
    plateau[tresor_pos[0]][tresor_pos[1]] = "T"

    # Étape 4 : Placer un piège aléatoire
    piege_pos = placer_piege(plateau, tresor_pos)
    plateau[piege_pos[0]][piege_pos[1]] = "X"

    # Étape 5 : Afficher le plateau
    afficher_plateau(plateau)

    # Étape 6 : Le joueur devine l'emplacement du trésor
    while True:
        devine_pos = demander_position(n, m, "Devinez la position du trésor (ligne colonne): ")
        if devine_pos == tresor_pos:
            print("Félicitations ! Vous avez trouvé le trésor !")
            break
        elif devine_pos == piege_pos:
            print("Oh non, vous êtes tombé dans un piège !")
            break
        else:
            print("Rien ici ! Essayez encore.")

if __name__ == "__main__":
    main()
```

```python
Exercice 2

def calculate_bmi(poids, taille):
    imc = poids / (taille ** 2)
    return round(imc, 2)

def categoriser_imc(imc):
    if imc < 18.5:
        return "Insuffisance pondérale"
    elif 18.5 <= imc < 25:
        return "Poids normal"
    elif 25 <= imc < 30:
        return "Surpoids"
    else:
        return "Obésité"

def convertir_poids_livres_en_kg(poids_livres):
   return poids_livres / 2.20462

def convertir_taille_pouces_en_metres(taille_pouces):
    """Convertit la taille de pouces en mètres."""
    return taille_pouces / 39.3701

def main():
    # Demande à l'utilisateur s'il souhaite entrer les valeurs en kilogrammes/mètres ou en livres/pouces
    systeme_mesure = input("Voulez-vous entrer le poids et la taille en kilogrammes et mètres (1) ou en livres et pouces (2)? Entrez 1 ou 2: ")

    if systeme_mesure == "1":
        poids = float(input("Entrez votre poids en kilogrammes: "))
        taille = float(input("Entrez votre taille en mètres: "))
    elif systeme_mesure == "2":
        poids_livres = float(input("Entrez votre poids en livres: "))
        taille_pouces = float(input("Entrez votre taille en pouces: "))
        poids = convertir_poids_livres_en_kg(poids_livres)
        taille = convertir_taille_pouces_en_metres(taille_pouces)
    else:
        print("Choix invalide.")
        return

    # Calcul de l'IMC
    imc = calculate_bmi(poids, taille)

    # Catégorisation de l'IMC
    categorie = categoriser_imc(imc)

    # Affichage du résultat
    print(f"Votre IMC est de {imc}. Vous êtes dans la catégorie : {categorie}.")

if __name__ == "__main__":
    main()

```

```python
Exercice 3
 import string

def nettoyer_chaine(input_string: str) -> str:
    """Nettoie la chaîne d'entrée en supprimant les caractères non alphanumériques et en convertissant en minuscules."""
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())
    return cleaned_string

def is_palindrome_while(input_string: str) -> bool:
    """Vérifie si la chaîne est un palindrome en utilisant une boucle while."""
    cleaned_string = nettoyer_chaine(input_string)
    left, right = 0, len(cleaned_string) - 1

    while left < right:
        if cleaned_string[left] != cleaned_string[right]:
            return False
        left += 1
        right -= 1

    return True

def is_palindrome_for(input_string: str) -> bool:
    """Vérifie si la chaîne est un palindrome en utilisant une boucle for."""
    cleaned_string = nettoyer_chaine(input_string)
    n = len(cleaned_string)

    for i in range(n // 2):
        if cleaned_string[i] != cleaned_string[n - i - 1]:
            return False

    return True

def tester_palindromes():
    exemples = [
        "A man, a plan, a canal, Panama",
        "No 'x' in Nixon",
        "Was it a car or a cat I saw?",
        "Hello, World!",
        "Madam, in Eden, I’m Adam",
        "12321",
        "123456"
    ]

    print("Test des fonctions is_palindrome_while et is_palindrome_for:")
    for exemple in exemples:
        print(f"Phrase: '{exemple}'")
        print(f"is_palindrome_while: {is_palindrome_while(exemple)}")
        print(f"is_palindrome_for: {is_palindrome_for(exemple)}\n")

if __name__ == "__main__":
    tester_palindromes()


```
